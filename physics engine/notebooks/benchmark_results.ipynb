# Harmonic Oscillator Benchmark (results)

This notebook contains the benchmark summary and a helper cell that runs the existing benchmark script. The numeric results were captured during a local run and reproduced here.

```python
# Re-run the benchmark script and print output (requires venv Python path)
import subprocess
cmd = ['"/Volumes/USB-HDD/coding projects/elements table/.venv/bin/python"', 'physics engine/examples/benchmark_harmonic.py']
try:
    out = subprocess.run(cmd, check=True, capture_output=True, text=True)
    print(out.stdout)
except Exception as e:
    print('Could not run benchmark script:', e)
```

```python
# Plot convergence results (requires matplotlib)
from pathlib import Path
p = Path('physics engine/examples/benchmark_convergence_results.csv')
if p.exists():
    import pandas as pd
    import matplotlib.pyplot as plt
    df = pd.read_csv(p)
    Ns = df['N'].values
    levels = (len(df.columns) - 1) // 3
    plt.figure(figsize=(6,4))
    for lvl in range(levels):
        plt.plot(Ns, df[f'level_{lvl}_relerr'], marker='o', label=f'level {lvl}')
    plt.xscale('log')
    plt.yscale('log')
    plt.xlabel('Basis size N')
    plt.ylabel('Relative error')
    plt.title('Convergence of HO eigenvalues vs analytic')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
else:
    print('CSV not found; run the convergence benchmark first')
```
